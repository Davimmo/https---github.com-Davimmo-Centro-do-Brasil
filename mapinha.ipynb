```python
import cv2
import numpy as np

# Carregar a imagem
imagem = cv2.imread('mapa.jpg')

# Converter a imagem para escala de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Detectar bordas
bordas = cv2.Canny(imagem_cinza, 50, 150)

# Encontrar contornos
contornos, _ = cv2.findContours(bordas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Definir o tamanho do quadrado
tamanho_quadrado = 50  # Substitua pelo tamanho desejado

# Lista para armazenar as coordenadas dos centros dos quadrados
centros = []

# Iterar sobre a imagem em etapas do tamanho do quadrado
for y in range(0, imagem.shape[0], tamanho_quadrado):
    for x in range(0, imagem.shape[1], tamanho_quadrado):
        # Verificar se o quadrado está inteiramente dentro da fronteira do país
        if cv2.pointPolygonTest(contornos[0], (x + tamanho_quadrado // 2, y + tamanho_quadrado // 2), False) == 1:
            # Adicionar as coordenadas do centro do quadrado à lista
            centros.append((x + tamanho_quadrado // 2, y + tamanho_quadrado // 2))
            # Colorir o quadrado de azul na imagem original
            imagem[y:y+tamanho_quadrado, x:x+tamanho_quadrado] = (255, 0, 0)  # BGR

# Calcular a média das coordenadas dos centros dos quadrados
media_x = int(np.mean([coord[0] for coord in centros]))
media_y = int(np.mean([coord[1] for coord in centros]))

# Colorir o ponto médio de vermelho na imagem original
cv2.circle(imagem, (media_x, media_y), 5 // 2, (0, 0, 255), -1)  # BGR

# Salvar a imagem de saída
cv2.imwrite('sua_imagem_saida2.png', imagem)
```

```python
import cv2
import numpy as np
def gerar_mapa(iteração,tamanho_quadrado):
    # Carregar a imagem
    imagem = cv2.imread('mapa.jpg')

    # Converter a imagem para escala de cinza
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    # Detectar bordas
    bordas = cv2.Canny(imagem_cinza, 50, 150)

    # Encontrar contornos
    contornos, _ = cv2.findContours(bordas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Definir o tamanho do quadrado
    # tamanho_quadrado = 50  # Substitua pelo tamanho desejado

    # Lista para armazenar as coordenadas dos centros dos quadrados
    centros = []

    # Iterar sobre a imagem em etapas do tamanho do quadrado
    for y in range(0, imagem.shape[0], tamanho_quadrado):
        for x in range(0, imagem.shape[1], tamanho_quadrado):
            # Verificar se o quadrado está inteiramente dentro da fronteira do país
            if cv2.pointPolygonTest(contornos[0], (x + tamanho_quadrado // 2, y + tamanho_quadrado // 2), False) == 1:
                # Adicionar as coordenadas do centro do quadrado à lista
                centros.append((x + tamanho_quadrado // 2, y + tamanho_quadrado // 2))
                
                # Desenhar o quadrado com borda preta
                cv2.rectangle(imagem, (x, y), (x + tamanho_quadrado, y + tamanho_quadrado), (0, 0, 0), 2)  # BGR
                
                # Colorir o quadrado de azul na imagem original
                imagem[y:y + tamanho_quadrado, x:x + tamanho_quadrado] = (255, 0, 0)  # BGR


    # Calcular a média das coordenadas dos centros dos quadrados
    media_x = int(np.mean([coord[0] for coord in centros]))
    media_y = int(np.mean([coord[1] for coord in centros]))

    # Colorir o ponto médio de vermelho na imagem original
    cv2.circle(imagem, (media_x, media_y), 8 // 2, (0, 0, 255), -1)  # BGR

    # Salvar a imagem de saída
    cv2.imwrite(f'sua_imagem_saida_{iteração}.png', imagem)

for i,tamanho in enumerate(range(201,1,-1)):
    gerar_mapa(i,tamanho)

```
